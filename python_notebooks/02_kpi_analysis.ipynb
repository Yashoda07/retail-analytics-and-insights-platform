```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/cleaned/Global_Superstore_cleaned.csv", sep=",", encoding="latin1")   #load the cleaned dataset, not the raw one
```

```python
df.head()
```

# KPI Layer - core metrics 

```python
# Calculate total sales

total_sales = df["sales"].sum()
print(f"Total Sales: ${total_sales:,.2f}")
```

```python
# Calculate total profit

total_profit = df["profit"].sum()
print(f"Total Profit: ${total_profit:,.2f}")
```

```python
# Calculate total orders

total_orders = df["order_id"].nunique()
print(f"Total Orders: {total_orders:,}")

# row_id = line item
# order_id = actual customer order
```

```python
# Calculate total customers

total_customers = df["customer_id"].nunique()
print(f"Total Customers: {total_customers:,}")
```

```python
# Calculate total quantity sold

total_quantity = df["quantity"].sum()
print(f"Total Quantity Sold: {total_quantity:,}")
```

```python
# Calculate average discount

avg_discount = df["discount"].mean()
print(f"Average Discount: {avg_discount:.2%}")
```

```python
# Calculate average delivery time

avg_delivery = df["delivery_days"].mean()
print(f"Average Delivery Time: {avg_delivery:.2f} days")
```

```python
# Calculate average profit margin

avg_profit_margin = df["profit_margin"].mean()
print(f"Average Profit Margin: {avg_profit_margin:.2f}%")
```

```python
# Create a summary of the KPIs

kpi_summary = {
    "Total Sales": total_sales,
    "Total Profit": total_profit,
    "Total Orders": total_orders,
    "Total Customers": total_customers,
    "Total Quantity": total_quantity,
    "Average Discount": avg_discount,
    "Average Delivery Days": avg_delivery,
    "Average Profit Margin": avg_profit_margin
}

pd.Series(kpi_summary)
```

```python
# Create a DataFrame for the KPIs

kpi_df = pd.DataFrame({
    "KPI": [
        "Total Sales",
        "Total Profit",
        "Total Orders",
        "Total Customers",
        "Total Quantity",
        "Average Discount",
        "Average Delivery Days",
        "Average Profit Margin"
    ],
    "Value": [
        total_sales,
        total_profit,
        total_orders,
        total_customers,
        total_quantity,
        avg_discount,
        avg_delivery,
        avg_profit_margin
    ]
})

kpi_df
```

```python
# Save the KPI DataFrame to a CSV file

kpi_df.to_csv("../data/cleaned/kpi_summary.csv", index=False)
```

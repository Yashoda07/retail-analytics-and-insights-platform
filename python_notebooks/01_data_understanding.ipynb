```python
# convert column names to lowercase (snake_case) and replace spaces with underscores 
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
      .str.replace("-", "_")
)
```

```python
#finding unique product IDs in the dataset
df['product_id'].nunique()
```

```python
# mixed date formats in the order_date and ship_date columns, so we will convert them to datetime format using pd.to_datetime() function with errors='coerce' parameter to handle any invalid date formats.

df["order_date"] = pd.to_datetime(
    df["order_date"],
    format="mixed",
    errors="coerce"
)

df["ship_date"] = pd.to_datetime(
    df["ship_date"],
    format="mixed",
    errors="coerce"
)
```

```python
# checks for missing values in the order_date column
df[df["order_date"].isna()]
```

```python

# checks for missing values in the ship_date column
df[df["ship_date"].isna()]
```

```python
# converting the data types of the columns to appropriate types

# string column to integer column
int_columns = [
    "row_id",
    "quantity",
    "postal_code"
]
df[int_columns] = df[int_columns].astype("int64")

# string column to float column
float_columns = [ 
    "sales",
    "discount",
    "profit"
]
df[float_columns] = df[float_columns].astype("float64")

# string column for date to datetime column
date_columns = [
    "order_date",
    "ship_date"
]
for col in date_columns:
    df[col] = pd.to_datetime(df[col], dayfirst=True)
```

```python
category_columns = [
    "order_id",
    "ship_mode",
    "customer_id",
    "customer_name",
    "segment",
    "country",
    "city",
    "state",
    "region",
    "product_id",
    "category",
    "sub_category",
    "product_name"
]

df[category_columns] = df[category_columns].astype("category")
```

----------------------------------------------- Feature Engineering ---------------------------------------------------------

```python
# extra column - 1, delivery_days

# calculating the delivery days for each order 
df["delivery_days"] = (
    df["ship_date"] - df["order_date"]
).dt.days
```

```python
df["delivery_days"].describe()
```

```python
#extra column - 2, profit_margin

# calculating the profit margin for each order
df["profit_margin"] = np.where(
    df["sales"] != 0,
    (df["profit"] / df["sales"]) * 100,
    0
)
```

```python
df["profit_margin"].describe()
```

```python
# extra column - 3, order_year

# calculating the order year for each order
df["order_year"] = df["order_date"].dt.year
```

```python
df["order_year"].describe()
```

```python
# extra column - 4, order_month

# calculating the order month for each order
df["order_month"] = df["order_date"].dt.month_name()
```

```python
df["order_month"].describe()
```

```python
# extra column - 5, order_quarter

# calculating the order quarter for each order
df["order_quarter"] = df["order_date"].dt.quarter
```

```python
df["order_quarter"].describe()
```

```python
# extra column - 6, order_weekday

# calculating the order weekday for each order
df["order_weekday"] = df["order_date"].dt.day_name()
```

```python
df["order_weekday"].describe()
```

```python
# checking for missing values in the newly created columns
new_columns = [
    "delivery_days",
    "profit_margin",
    "order_year",
    "order_month",
    "order_quarter",
    "order_weekday"
]

df[new_columns].isnull().sum()
```
